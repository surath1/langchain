```python
from langchain_openai import ChatOpenAI
from constants import OPENAI_KEY
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain

llm = ChatOpenAI(openai_api_key=OPENAI_KEY, temperature=0.7)
```

```python
def generate_pet_name(comp_type, location):    
    prompt_name = PromptTemplate(
        input_variables = ['comp_type','location'],
        template = "please suggest two cool name of my {comp_type} company name and its address in {location}."
    )

    name_chain = LLMChain(llm=llm, prompt=prompt_name, output_key="location")

    response = name_chain({'comp_type': comp_type, 'location': location})


    return response

if __name__ == "__main__":
    print(generate_pet_name("Software", "USA"))
```
